```python
#scatter plot
# it is used to scatter x and y value
#plot 2d graph
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('scatter plot')
plt.show()
```

```python
y=x*x
```

```python
# plt plot
plt.plot(x,y,'g')
```

```python
#if i want multiple dash line in this
plt.plot(x,y,'g--')
```

```python
#if i want some other dash line in this
plt.plot(x,y,'r*-')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('plot')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
```

```python
x=np.arange(0,4,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```
